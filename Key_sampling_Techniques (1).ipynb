## Systematics Sampling

```python
def system_sample(data_frame,step_size):
    index_selected=np.arange(0,data_frame.shape[0],step=step_size)
    df_new=data_frame.iloc[index_selected]
    return df_new
```

```python
np.arange(0,10,5) #give you specific no. of elements in the range for a given step size(interval)
```

```python
np.linspace(0,10,5) # give you the quantity of elements that you specify: uniform distance from one another
```

```python
# call the systematic sampling
system_sample(housing,15)
```
